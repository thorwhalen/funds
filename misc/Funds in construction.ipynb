````python
from itertools import cycle
from functools import reduce, lru_cache
from operator import and_, attrgetter
from i2 import Pipe

from funds.scrap.company_info_w_historical_metrics import get_simfin_src_store

def get_src_store(src_store=None):
    if src_store is None:
        return get_simfin_src_store()
    else:
        return src_store
    
def df_info(name, z=None):
    z = get_src_store(z)
    df = z[name]
    print(name + '\n')
    print(f"{df.shape=}, {df.Ticker.nunique()=}\n")
    print(df.iloc[0])

names = [
    'us-derived-banks-quarterly.csv',
    'us-derived-insurance-quarterly.csv',
    'us-derived-quarterly.csv',
    'us-balance-banks-quarterly-full.csv',
    'us-balance-insurance-quarterly-full.csv',
    'us-balance-quarterly-full.csv',
    'us-cashflow-banks-quarterly-full.csv',
    'us-cashflow-insurance-quarterly-full.csv',
    'us-cashflow-quarterly-full.csv',
    'us-income-banks-quarterly-full.csv',
    'us-income-insurance-quarterly-full.csv',
    'us-income-quarterly-full.csv'
]
    
data_groups = {
    "banks": (
        "us-derived-banks-quarterly.csv",
        "us-balance-banks-quarterly-full.csv",
        "us-cashflow-banks-quarterly-full.csv",
        "us-income-banks-quarterly-full.csv"
    ),
    "insurance": (
        "us-derived-insurance-quarterly.csv",
        "us-balance-insurance-quarterly-full.csv",
        "us-cashflow-insurance-quarterly-full.csv",
        "us-income-insurance-quarterly-full.csv",
    ),
    "rest": (
        "us-derived-quarterly.csv",
        "us-balance-quarterly-full.csv",
        "us-cashflow-quarterly-full.csv",
        "us-income-quarterly-full.csv"
    )
}


def get_names(names, names_for_group=data_groups):
    if isinstance(names, str):
        group = names
        return names_for_group[group]
    else:
        return names
    
def get_dfs(dfs, z=None):
    z = get_src_store(z)
    if not isinstance(next(iter(dfs)), pd.DataFrame):
        group_or_names = dfs
        names = get_names(group_or_names)
        dfs = tuple(map(z.__getitem__, names))
    return dfs

def common_cols(dfs):
    return tuple(reduce(and_, tuple(map(Pipe(attrgetter('columns'), set), dfs))))

def analyze_group(dfs, z=None):
    z = get_src_store(z)
    dfs = get_dfs(dfs, z)
    d = {}
    d['common_cols'] = common_cols(dfs)
    d['shapes'] = list(map(lambda x: x.shape, dfs))
    merged = pd.concat(dfs, axis=1)
    d['merged_shape'] = merged.shape
    return d

def merge_group(dfs, keys=None, z=None):
    z = get_src_store(z)
    dfs = get_dfs(dfs, z)
    if keys is None:
        keys = common_cols(dfs)
    merged = pd.concat(
        dfs, 
        keys=keys, 
        axis=0,
        ignore_index=True,
        verify_integrity=True,
    )
    merged = pd.merge()
    assert len(set(merged.columns)) == len(merged.columns), "columns are not unique"
    return merged

analyze_group(data_groups['banks'])
````

````python
# Get most simular minfo entry!

from functools import lru_cache
from grub import SearchStore
from sklearn.metrics.pairwise import cosine_similarity
from itertools import chain


def transform_text(text, transformer):
    if isinstance(text, str):
        return transformer([text])[0]
    else:
        return transformer(text)
    
@lru_cache
def mk_text_to_vect(*learn_texts):

    docs = dict(enumerate(chain(*learn_texts)))
    s = SearchStore(docs)
    return partial(transform_text, transformer=s.tfidf.transform)


def max_semantically_simular(
    ticker='AAPL', target=info['description'], src=minfo['longBusinessSummary'], text_to_vect=None
):
    if ticker not in src:
        return None
    
    text_to_vect = text_to_vect or mk_text_to_vect(tuple(target.dropna()), tuple(src.dropna()))
        
    src_to_search = src.iloc[src.index.isin(target.index)].dropna()
    
    description = target.loc[ticker]
    if description:
        idx = cosine_similarity(
            text_to_vect([target.loc[ticker]]),
            text_to_vect(src_to_search)
        ).argmax()
        return src_to_search.index[idx]
        
        

def max_semantically_simular(
    ticker='AAPL', target=info['description'], src=minfo['longBusinessSummary'], text_to_vect=None
):
    if ticker not in src:
        return None
    
    text_to_vect = text_to_vect or mk_text_to_vect(tuple(target.dropna()), tuple(src.dropna()))
        
    src_to_search = src.iloc[src.index.isin(target.index)].dropna()
    
    description = target.loc[ticker]
    if description:
        idx = cosine_similarity(
            text_to_vect([target.loc[ticker]]),
            text_to_vect(src_to_search)
        ).argmax()
        return src_to_search.index[idx]
    
#         return src_df.iloc[idx]['ticker'] == ticker
````

````python
from guise import word_cloud, url_to_html, html_tokens
from i2 import Pipe
from collections import Counter
from urllib.parse import urljoin

from bs4 import BeautifulSoup

def html_to_links(html):
    soup = BeautifulSoup(html, "lxml")
    for link in soup.findAll('a'):
        yield link.get('href')
        
def tokens_from_url(url, max_n_links=7):
    html = url_to_html(url)
    yield from html_tokens(html)
    top_links = Counter(html_to_links(html)).most_common(n=max_n_links)
    for link, _ in top_links:
        try:
            if not link.startswith('http'):
                link = urljoin(url, link)
#             print(link)
            _html = url_to_html(link)
            yield from html_tokens(_html)
        except Exception as e:
            print(f"Error: {type(e).__name__}: {e}")
            

ticker_to_homepage_html = Pipe(website_of_ticker.__getitem__, url_to_html)
ticker_to_tokens = ticker_to_homepage_html = Pipe(website_of_ticker.__getitem__, tokens_from_url)
ticker_to_word_cloud = Pipe(ticker_to_tokens, word_cloud)
````
